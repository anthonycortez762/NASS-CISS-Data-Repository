```python
import pandas as pd

ciss = pd.read_csv("./CISS/clean/INJURY_CLEANED.csv")

dict = {}

for year in ciss['YEAR'].unique():
    annual_ciss = ciss[ciss['YEAR']==year]
    dict[year] = annual_ciss['INJLEVEL'].value_counts().reset_index()

print(dict)
```
